1. first half of Ascending Order and second half Descending order


```python
arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
rev = []
for i in range(0, round(len(arr) / 2) + 1):
    rev.append(arr[i])
for j in range(len(arr) - 1, round(len(arr) / 2), -1):
    rev.append(arr[j])

print(rev)
```

2. Take an array start at middle and move left to right at a time

Steps: 1.Take new arr and store middle value at start and later loop left when left is greater than 0 and right while right is less than input length.Print middle and left,right values at a time using 2 pointers


```python
arr = [1, 2, 3, 4, 5, 6, 7, 8]
middleValue = round(len(arr) / 2)
left = round(len(arr) / 2)
right = round(len(arr) / 2)
newArr = [
    arr[middleValue],
]

while left > 0 or right < len(arr) - 1:
    if left > 0:
        left += -1
        newArr.append(arr[left])
    if right < len(arr) - 1:
        right += 1
        newArr.append(arr[right])

print(newArr)
```

4. Swap elements based on values of adjecent values


```python
array = [10, 1, 20, 3, 4, 5, 6]

for i in range(len(array) - 1):
    for j in range(len(array) - 1):
        if array[j] > array[j + 1]:
            temp = array[j]
            array[j] = array[j + 1]
            array[j + 1] = temp


print(array)
```

Example: Output: [11, 12, 22, 25, 34, 64, 90]
